```python
# Magic
%matplotlib inline

# Libraries in use
import os
import sys
import math
import pylab as pl
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from sklearn import linear_model, neural_network, metrics
from matplotlib import collections  as mc

from myutil import *
# My libraries
#sys.path.append('../infra/')
#from AssetsData import AssetsData
```

```python
# Read the data
df = pd.read_csv("../../data/train.csv.gz")
```

```python
ids = sorted(set(df['ID']))
```

```python
# Use self-written wrapper for quicker data extraction
data  = AssetsData()
#scores = []
assetslist = {}
#number_of_assets = 1315

for asset_name in ids:
    
    #asset_name = 'S' + '{:0>4}'.format(i)
    #print asset_name
    asset_df = data.get_asset_df(asset_name, df) #TODO some feature preprocess
    #print asset_df.isnull().sum()
    y, df_without_nan = data.get_y_and_factors(asset_df)
    x = list(np.array(df_without_nan))
    #print df_without_nan.isnull().sum()
    
    asset = Assets(asset_name)
    
    if len(y) < 300:
        assetslist[asset_name] = asset
        continue
    
    a = np.array(x)
    stds = np.std(a, axis=0)
    l1 = (stds > 0.000001) 
    l2 = np.isnan(stds)
    l3 = (stds < 10) 
    bl = [l1[i] and l3[i] and not l2[i] for i in range(len(l1))]
    #print bl
    
    asset.factors = df_without_nan.columns[bl]
    
    #print a.shape
    #print stds != 0
    #print df_without_nan.columns
    #print df_without_nan.columns[stds != 0]

    a_trim_const = a[:, bl]
    a_winsorize = scipy.stats.mstats.winsorize(a_trim_const, limits = (.1, .1), axis=0)

    asset.mean = np.mean(a_winsorize, axis=0)
    asset.var = np.std(a_winsorize, axis=0) 
    a_zero_mean_unit_var = (a_winsorize - np.mean(a_winsorize, axis=0))/np.std(a_winsorize, axis=0)

    #mutual_infos = np.array([metrics.normalized_mutual_info_score(y, a_zero_mean_unit_var[:, i]) for i in range(a_zero_mean_unit_var.shape[1])])                         
    #print mutual_infos
    #print asset.var
    #print asset.factors

    X = list(a_zero_mean_unit_var)
    
    clf = linear_model.Lasso(alpha=0.000001, max_iter=10000)
    clf.fit(X,y)
    
    for i in clf.coef_:
        if i > 100:
            raise('too large')
    
    asset.model = clf
    asset.score = clf.score(X, y)
    asset.day_count = len(y)
    
    assetslist[asset_name] = asset
    #scores.append(clf.score(X,y))
    
    print asset.name
    #print asset.name, asset.factors, asset.model, asset.mean, asset.var
```
